```python
#返回初步处理过的email list
def readfile(filename):
    emaillist = []
    f = open(filename,encoding='utf8')
    for line in f.readlines():
        linestr = line.strip()
        linestrlist = linestr.split(" ")
        emaillist.append(linestrlist)
    return emaillist

# def readstopwords(filename):
#     stopwordslist = []
#     f = open(filename,encoding='utf8')
#     for line in f.readlines():
#         linestr = line.strip()
#         stopwordslist.append(linestr)
#     return stopwordslist


#提取正负样本
def fileextraction(filename):
    #记录label0的邮件个数
    label0count = 0
    label1count = 0
    #存储label0的邮件的words分布
    label0dict = {}
    label1dict = {}
    emaillist = readfile(filename)
    for email in emaillist:
        emailset = set(email[1:])
        #print(emailset)
        if email[0] == '0':
            label0count += 1
            for word in emailset:
                if word in label0dict:
                    label0dict[word] += 1
                else:
                    label0dict[word] = 1
        if email[0] == '1':
            label1count += 1
            for word in emailset:
                if word in label1dict:
                    label1dict[word] += 1
                else:
                    label1dict[word] = 1

    
    return [label0count,label0dict, label1count, label1dict]
```

```python
trainfilename = "E:/study/第四年/知识分析/Naive_Bayes/NB/spam_train.txt"
testfilename = "E:/study/第四年/知识分析/Naive_Bayes/NB/spam_test.txt"
returnlist = fileextraction(trainfilename)
#stop_words = "E:/study/data/停用词表stopwords/英文.txt"
#stop_words_list = readstopwords(stop_words)
#print(stop_words_list)
```
