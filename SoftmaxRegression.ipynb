```python
import numpy as np
import pandas as pd
import math
```

```python
# softmax　regression的思路
# softmax regression是logistic regression在多分类上的拓展
# 首先，要给出预测函数，因为我们需要做的事情是对不同的样本，给出其正确的分类，所以我们的预测函数即K个类别对应的估计概率值。
# htheta(x)=[P(y=1|x,theta),
#           p(y=2|x,theta),
#           ……
#           P(y=k|x,theta)]
#          =e/sum(e)
# 第二步，得到了预测函数，我们需要找到合适的损失函数(合适的概念是指函数最好是凸的或者有解析解容易得到最小值的)，使得损失函数最小,这里使用了
# 极大似然估计。
# 极大似然估计：也就相当于argmax,求如何取参数，使得最后可以得到的结果最好
# 代价函数：J(theta)=-[m(sum)i=1K(sum)i1{y(i)=k}log(e/(sum)e)] 公式没法敲，意会一下 很多资料都有这个公式
# 对其求导，使用梯度下降求theta。这里就单纯得使用了全局梯度下降
# softmax 和 logistic regression回归的难点在于如何理解预测函数，损失函数。
# 预测函数就是我们的目的：分类。虽然这俩叫regression 但实际上是分类问题。
# 损失函数就是代价，分错的时候，这个代价越大越好。以此为目标进行构建，常见的有方差，似然估计等等。
# 那么问题来了，为什么logistic regression要叫做logistic呢？
# 就是因为预测函数使用了sigmoid函数。
# 我们再次回到我们的目的上，预测函数是为了分类。那么也就是说，对于不同的类别，预测函数要具有很高的区分度，这样才能方便损失函数区别，最终才能提高准确率。
# 那么sigmoid就不失为一个很好的选择了。我们知道，在x=0的时候，sigmoid的导数是很大的，这个性质使得只要我们的theta将两类分到了x<0和x>0两类，
# 我们的p就会有很大的一个差距。看softmax的代价函数我们知道，当y==k但我们的p(k)很小时，J就会很大，这样是不好的。因此我们要做的是y==k时尽量让这时候的p最大。
```

```python
def train(k,n,train_data):
    alpha = 0.1#学习率
    epoch = 200#迭代次数
    m = len(train_data)#m代表数据总条数
    
    theta = generatefeaturematrix(k,n)
    x_train = (np.matrix(train_data.iloc[:,0:n]))
    y_train = []
    for i in train_data[n]:
#         #print(i)
#         if i == 'Iris-setosa':
#             y_train.append(0)
#         if i == 'Iris-versicolor':
#             y_train.append(1)
#         if i == 'Iris-virginica':
#             y_train.append(2)  
        y_train.append(i)
    #print(theta)
    #print(y_train)
    #print(x_train)
    #print(x_train[0].shape)
    #print((theta.T)[0])
    for item in range(epoch):
        print("epoch:%d================>"%item)
        print(theta)
        J = generatefeaturematrix(k,n)
        for itemk in range(k):
            #print("itemk:",itemk)
            temp = 0
            for i in range(m):
                #print("itemk:",itemk)
                if itemk == y_train[i]-1:
                    temp += x_train[i]*(1-generateP(theta,x_train[i].T,k)[itemk])
                else:
                    temp += x_train[i]*(0-generateP(theta,x_train[i].T,k)[itemk])
            # print("temp",temp)
            J[:,itemk] = -temp/m
        #print("J:",J)
        theta = theta-alpha*J
    return theta 
    
    
def generateP(theta,x,k):
    P = []
    normal = 0
    for i in range(k):
        P.append(math.exp(np.linalg.det((theta.T)[i]*x)))
        normal += math.exp(np.linalg.det((theta.T)[i]*x))
    for i in range(k):
        #print(i)
        P[i] = P[i]/normal
    #print("P:",P)
    return P
```

```python
theta = train(3,13,train_data)
```
