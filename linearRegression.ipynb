```python
def lwlr(testPoint,xArr,yArr,k=1):
    #print("testPoint:",testPoint)
    xMat = np.mat(xArr)
    yMat = yArr
    #print("yMatshape:",yMat.shape)
    xt = xMat.T
    m = np.shape(xMat)[0]
    weights = np.mat(np.eye((m)))                                        #创建权重对角矩阵
    for j in range(m):                                                  #遍历数据集计算每个样本的权重
        diffMat = testPoint - xMat[j, :]                                 
        weights[j, j] = np.exp(diffMat * diffMat.T/(-2.0 * k**2))

    para = lg.inv(xt * weights * xMat)
    para = para.dot(xt)
    para = para.dot(weights)
    para = (para.dot(yMat)).T
    #para = ((np.transpose(xt.dot(xtrain))).dot(xt)).dot(ytrain)
    #print(para)
    #print(xtrain.dot(para))
    return testPoint * para
```

```python
def lwlrEvaluate(xValid,yValid,xArr,yArr,k):
    bestk = 0
    minmse = 999999
    for item in k:
        mse = 0
        yPred = lwlrTest(xValid,xtrain,ytrain)
        for i in range(len(yPred)):
            mse += (yPred[i] - yValid[i])**2
        if mse < minmse:
            minmse = mse
            bestk = item
            print(minmse)
    return bestk
```

```python
k = []
for i in range(0,50):
    k.append(i/50)
#print(k)
bestk = lwlrEvaluate(xvalid,yvalid,xtrain,ytrain,k)
print(bestk)
```

```python
def lwlrTest(testArr,xArr,yArr,k=1.0):
    #print("yArr's shape:",yArr.shape)
    testMat = np.mat(testArr)
    #print("testArr：",testArr)
    m = np.shape(testArr)[0]                                            #计算测试数据集大小
    print("m:",m)
    #print("testpoint's type:",type(testArr))
    #print("testPoint[0]:",testArr.iloc[0])
    yHat = np.zeros(m)    
    for i in range(m):                                                    #对每个样本点进行预测
        yHat[i] = lwlr(testMat[i],xArr,yArr,k)
        print("yHat[%d]:%f"%(i,yHat[i]))
    return yHat

xtest = test.iloc[:,0:8]
for i in xtest.index:
    if test.iloc[i,0] == "F":
        xtest.iloc[i,0] = -1
    elif test.iloc[i,0] == "M":
        xtest.iloc[i,0] = 1
    elif test.iloc[i,0] == "I":
        xtest.iloc[i,0] = 0
        
yHat = lwlrTest(xtest,xtrain,ytrain,0.25)
print(yHat)
```
